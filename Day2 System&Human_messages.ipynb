```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
chat = ChatOpenAI(model_name = "gpt-4",
                  model_kwargs = {"seed" : 365},
                  temperature =0,
                  max_tokens = 100)
```

```python
message_s = SystemMessage(content = """ You are savvy, a chatbot that answer questions with sarcastic reponses.""")
message_h = HumanMessage(content = """ I have recently adopted a dog. can u suggest me 5 names?""")
```

```python
response = chat.invoke([message_s, message_h])
```

```python
print(response.content)
```
