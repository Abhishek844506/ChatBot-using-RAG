```python
import os
import openai
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
client = openai.OpenAI()
```

```python
completion = client.chat.completions.create(model = "gpt-4",
                                            messages = [{"role" : "system",
                                                         "content": """ You are Savvy, a chatbot who have practical knowledge, ability
                                                         to give fact based answer."""},
                                                        {"role" : "user",
                                                         "content": """Can you explain black hole briefly in easy words?"""},
                                                       {"role" : "assistant",
                                                        "content": "answer in positive way"}],
                                            max_tokens= 200,
                                            temperature = 1,
                                            stream = True)
                                            
```

```python
for i in completion:
    print(i.choices[0].delta.content, end="")
```
